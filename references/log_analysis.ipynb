```python
import sys
import re
import matplotlib.pyplot as plt
```

```python
temp_lst = []
train_mae_lst = []
epo_num_lst = []
val_mae_lst = []
```

```python
with open('/Users/xiangyikong/Desktop/Flu_preprocess/quantile_model_for_flu/Tweek15_unshuffled.txt', 'r') as f:
    for line in f:
        if "INFO - Epoch" in line:
            if "val_mae:" in line:
                temp_train_mae = line.split("train_mae:")[1].split("val_mae:")[0]
                temp_val_mae = line.split("val_mae:")[1].split(", lr:")[0]
                epoch = line.split("/")[0].split("[")[1]
                temp_lst.append(line)
                train_mae_lst.append(float(temp_train_mae.replace(",", "")))
                epo_num_lst.append(int(epoch))
                val_mae_lst.append(float(temp_val_mae))
```

```python
# naming the x axis
plt.xlabel('Number of Epochs')
# naming the y axis
plt.ylabel('MAE Loss')
  
# giving a title to my graph
plt.title('Training and Validation MAE over Epochs')
plt.plot(val_mae_lst[122:367], label="val_mae")
plt.plot(train_mae_lst[122:367], label="train_mae")
plt.legend()
plt.show()
```
